```python
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt

from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

import torch
from torch import nn
from torch.nn import functional as F

from torch.utils.data import TensorDataset, DataLoader
from scipy.special import softmax
from random import uniform
from math import cos, sin, pi

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

sns.set(style="darkgrid", font_scale=1.4)
```

```python
f = open("pursuit_of_profit_input.txt", "r")
```

```python
n = int(f.readline().split()[0])
n
```

```python
dots = np.empty((n, 2), dtype=int)
```

```python
for i in range(n):
    dots[i] = [int(i) for i in f.readline().split()]
```

```python
dots[-1]
```

```python
plt.scatter(dots[:,0],dots[:,1])
```

```python
clf = KNeighborsClassifier(metric='euclidean', n_neighbors=100, n_jobs=-1)
```

```python
clf.fit(dots, np.zeros(n))
```

```python
y = clf.predict(dots)
```

```python
y 
```

```python
plt.scatter(dots[:,0],dots[:,1], c = y)
```
